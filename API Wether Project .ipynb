```python
import requests
import pandas as pd 

# List of cities from various talukas/cities in Maharashtra

city_name = [
    "Mumbai", "Mumbai", "Andheri", "Borivali", "Malad", "Colaba",
    "Pune", "Pimpri-Chinchwad", "Haveli", "Mulshi", "Khed", "Maval","Aundh","Junnar",
    "Nagpur", "Kamptee", "Narkhed", "Umred", "Saoner",
    "Nashik", "Nandgaon", "Dindori", "Trimbak", "Igatpuri",
    "Aurangabad", "Jalna", "Paithan", "Sillod", "Kannad",
    "Kolhapur", "Hatkanangale", "Shirol", "Karveer", "Radhanagari",
    "Solapur", "Barshi", "Karmala", "Pandharpur", "Mangalvedha",
    "Thane", "Kalyan", "Ulhasnagar", "Ambernath", "Badlapur", "Bhiwandi",
    "Satara", "Karad", "Khatav", "Wai", "Phaltan",
    "Sawantwadi", "Kudal", "Malvan", "Vengurla",
    "Chandrapur", "Ballarpur", "Warora", "Rajura", "Bhivpur",
    "Beed", "Patoda", "Georai", "Kaij", "Ashti",
    "Latur", "Ausa", "Nilanga", "Deoni", "Renapur",
    "Malkapur", "Chikhli", "Khamgaon", "Shegaon", "Jalgaon Jamod",
    "Jalna", "Badnapur", "Partur", "Mantha", "Jafrabad",
    "Ratnagiri", "Chiplun", "Dapoli", "Rajapur", "Khed",
    "Hingoli", "Aundha", "Kalamnuri", "Sengaon",
    "Wardha", "Arvi", "Deoli", "Karanja",
    "Yavatmal", "Pusad", "Arni", "Wani", "Digras",
    "Akola", "Murtizapur", "Patur", "Baranjala", "Telhara",
    "Amravati", "Morshi", "Chandur Bazar", "Achalpur", "Nandgaon Khandeshwar",
    "Alibag", "Uran", "Panvel", "Karjat", "Khalapur",
    "Washim", "Khamgaon", "Manora", "Pusad"]

API_Key="1be6fe3a073779145b4ccfa2dad856ec"

# List to hold all weather data
weather_data_list = []

# Fetch data for each city
for city in city_name:
    try:
        Req=requests.get("https://api.openweathermap.org/data/2.5/weather?q={}&appid=1be6fe3a073779145b4ccfa2dad856ec&units=metric"
.format(city))
        if Req.status_code == 200:
            data = Req.json()

            # Extract relevant weather data
            weather_data = {
                "City": data["name"],
                "Temperature (°C)": data["main"]["temp"],
                "Min_Temperature (°C)": data["main"]["temp_min"],
                "Max_Temperature (°C)": data["main"]["temp_max"],
                "Feels_Like (°C)": data["main"]["feels_like"],
                "Pressure (hpa)": data["main"]["pressure"],
                "Weather": data["weather"][0]["description"],
                "Humidity (%)": data["main"]["humidity"],
                "Wind Speed (m/s)": data["wind"]["speed"],
                "TimeZone (sec)": data["timezone"]
            }
            
            # Add the weather data for the current city to the list
            weather_data_list.append(weather_data)
        else:
             print(f"Error: Could not fetch data for {city}")
    except Exception as e:
        print(f"An error occurred while fetching data for {city}: {e}")

                                    
df = pd.DataFrame(weather_data_list) # Convert the list of weather data to a DataFrame
print(df)

# Save the DataFrame to a CSV file
df.to_csv("maharashtra_weather_data3.csv", index=False)
```
